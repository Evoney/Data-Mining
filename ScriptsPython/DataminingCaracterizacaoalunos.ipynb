```python
#importação das bibliotecas
import matplotlib.pyplot
import re
import pandas as pd
import os
import seaborn as sns
```

```python
#Setando o local do arquivo
THIS_FILE_NAME = 'analise-correlação.py'
THIS_FILE_PATH = '/home/evoney/Documents/Pibic/' 


PLANILHA_carac_2018 = 'rawCaracterizacaoQPA-E.2018.csv'
PLANILHA_carac_2019 = 'rawCaracterizacaoQPA-E.2019.csv'

os.chdir(THIS_FILE_PATH)

df_carac_2018 = pd.read_csv(PLANILHA_carac_2018, sep=',')
df_carac_2019 = pd.read_csv(PLANILHA_carac_2019, sep=',')
```

```python
df_carac_2018_2019 = df_carac_2018.append(df_carac_2018)
```

```python
plt.figure(figsize=(12,8))
ax = sns.countplot(x="Como é o seu acesso à internet em casa?",data=df_carac_2018_2019, palette="Blues_d")
ncount = len(df_carac_2018_2019)
plt.title('Distribuição de alunos acerca do acesso à internet 2018 e 2019')
plt.xlabel(" ")


# Make twin axis
ax2=ax.twinx()

# Switch so count axis is on right, frequency on left
ax2.yaxis.tick_left()
ax.yaxis.tick_right()

# Also switch the labels over
ax.yaxis.set_label_position('right')
ax2.yaxis.set_label_position('left')

ax2.set_ylabel('Frequência [%]')
ax.set_ylabel('Quantidade')

for p in ax.patches:
    x=p.get_bbox().get_points()[:,0]
    y=p.get_bbox().get_points()[1,1]
    ax.annotate('{:.1f}%'.format(100.*y/ncount), (x.mean(), y), 
            ha='center', va='bottom') # set the alignment of the text

ax2.set_ylim(0,100)
ax.set_ylim(0,ncount)
ax2.grid(None)
plt.savefig('Distribuiçãointernet.png')
```

```python
plt.figure(figsize=(12,8))

df_equipamentos = pd.DataFrame(columns = ['equipamentos'])
lists = []
for i in df_carac_2018_2019["Quais destes equipamentos de informática você possui em casa?"]:
    if 'Computador (desktop ou notebook)' in i:
        lists.append('Computador (desktop ou notebook)')
    if 'Smartphone' in i:
        lists.append('Smartphone')
    if 'Tablet' in i:
        lists.append('Tablet')
    if 'Nenhum dos equipamentos acima' in i:
        lists.append('Nenhum dos equipamentos acima')

df_equipamentos['equipamentos'] = lists

ax = sns.countplot(x="equipamentos",data=df_equipamentos, palette="Blues_d")
ncount = len(df_carac_2018_2019)
plt.title('Distribuição acerca do uso de equipamentos de informática 2018 e 2019')
plt.xlabel(" ")


# Make twin axis
ax2=ax.twinx()

# Switch so count axis is on right, frequency on left
ax2.yaxis.tick_left()
ax.yaxis.tick_right()

# Also switch the labels over
ax.yaxis.set_label_position('right')
ax2.yaxis.set_label_position('left')

ax2.set_ylabel('Frequência [%]')
ax.set_ylabel('Quantidade')

for p in ax.patches:
    x=p.get_bbox().get_points()[:,0]
    y=p.get_bbox().get_points()[1,1]
    ax.annotate('{:.1f}%'.format(100.*y/ncount), (x.mean(), y), 
            ha='center', va='bottom') # set the alignment of the text

ax2.set_ylim(0,100)
ax.set_ylim(0,ncount)
ax2.grid(None)
plt.savefig('Distribuiçãocomputador.png')
```
